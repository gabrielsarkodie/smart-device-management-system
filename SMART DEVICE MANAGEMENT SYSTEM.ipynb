```


class SmartDevice:
    def __init__(self, name, device_id):
        self.name = name
        # Encapsulated attributes
        self.__device_id = device_id
        self.__power_status = False

    # Getters and setters
    @property
    def device_id(self):
        return self.__device_id

    @device_id.setter
    def device_id(self, new_id):
        if len(new_id) > 0:
            self.__device_id = new_id
        else:
            print("ID cannot be empty!")

    def turn_on(self):
        self.__power_status = True
        print(f"{self.name} is now ON.")

    def turn_off(self):
        self.__power_status = False
        print(f"{self.name} is now OFF.")

    def display_info(self):
        status = "ON" if self.__power_status else "OFF"
        print(f"Device: {self.name} | ID: {self.__device_id} | Status: {status}")

class TemperatureSensor(SmartDevice):
    def __init__(self, name, device_id, temperature=25):
        super().__init__(name, device_id)
        self.temperature = temperature

    def read_temperature(self):
        print(f"{self.name} reading: {self.temperature}°C")

class SmartLight(SmartDevice):
    def __init__(self, name, device_id, brightness=50):
        super().__init__(name, device_id)
        self.brightness = brightness

    def increase_brightness(self):
        if self.brightness < 100:
            self.brightness += 10
            print(f"Brightness increased to {self.brightness}%")
        else:
            print("Max brightness reached.")

    def decrease_brightness(self):
        if self.brightness > 0:
            self.brightness -= 10
            print(f"Brightness decreased to {self.brightness}%")

class SecurityCamera(SmartDevice):
    def __init__(self, name, device_id):
        super().__init__(name, device_id)
        self.recording = False

    def start_recording(self):
        self.recording = True
        print(f"{self.name} is now recording.")

    def stop_recording(self):
        self.recording = False
        print(f"{self.name} stopped recording.")

# Main menu loop
def main():
    # Creating instances
    temp_sensor = TemperatureSensor("Living Room Sensor", "TS101")
    light = SmartLight("Bedroom Light", "SL202")
    cam = SecurityCamera("Front Door Cam", "SC303")

    devices = [temp_sensor, light, cam]

    while True:
        print("\n--- Smart Device System ---")
        print("1. Display Info\n2. Turn On\n3. Turn Off\n4. Read Temp\n5. Adjust Light\n6. Camera Rec\n7. Exit")

        choice = input("Enter choice: ")

        if choice == '1':
            for d in devices:
                d.display_info()
        elif choice == '2':
            light.turn_on()
        elif choice == '3':
            light.turn_off()
        elif choice == '4':
            temp_sensor.read_temperature()
        elif choice == '5':
            light.increase_brightness()
        elif choice == '6':
            cam.start_recording()
        elif choice == '7':
            print("Exiting...")
            break
        else:
            print("Invalid choice, try again.")

if __name__ == "__main__":
    main()
```
